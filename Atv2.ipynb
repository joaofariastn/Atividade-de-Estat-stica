```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-100, 100, 1000)
y = (x**2 + 2*x + 1) * np.sin(x)
```

```python
df = pd.DataFrame({'x': x, 'y': y})
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['x'], df['y'], label='y = (x^2 + 2x + 1) * sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gr√°fico')
plt.grid(True)
plt.legend()
plt.show()
```
